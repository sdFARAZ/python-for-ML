### Q2) a)

```python
from scipy import linalg

a=np.array([[-4,-5,-1],[-2,-5,-2],[-2,5,2]])
b=np.array([18,12,15])
x=linalg.solve(a,b)
sol = [f"{value:.2f}" for value in x]
print('x =', sol[0])
print('y = ', sol[1])
print('z = ',sol[2])
```

### Q2) b)

```python
a=np.array([[4,4,1],[2,-4,1],[5,-4,-5]])
b=np.array([24,0,12])
sol=linalg.solve(a,b)
print('x =', sol[0])
print('y = ', sol[1])
print('z = ',sol[2])
```

### Q3) a)

```python
a=np.array(([3,2,7],[1,0,4],[4,6,9]))
vector,value=linalg.eig(a) #linalg.eig gives two outputs for values and vectors 
print('Eigenvector = ',vector)
print(' ')
print(f'Eigenvalues:\n{value}')
print(' ')
i=linalg.inv(a)
print(f'Inverse of the matrix is:\n{i}')
print(' ')
x=linalg.det(a)
print('Determinant of the matrix = ', x)
```
